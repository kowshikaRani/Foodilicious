```python
# Necessary imports
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import os
import json
import time
import spoonacular as sp
```

```python
food_url = 'https://foodista.com/community-recipes'

# Request to retrive data from URL
food_response = requests.get(food_url)

# Parsing the whole HTML page using BeautifulSoup
food_soup = BeautifulSoup(food_response.text, 'html.parser')
print(food_soup)
```

```python
# Getting the recipes in the first page
span = food_soup.find_all('span', {'class': 'field-content'})
recp = []
for ele in span:
  for value in ele.find_all('a'):
    dish=value.get('href')
    print(dish)
    recp.append(dish)
```

```python
# Find all a tags with class active to find the necessary pages
a = food_soup.find_all('a', {'class': 'active'})
print(a)
```

```python
# Getting the relative path of all recipes from all pages
unique_pages=set()
c = 0
for tag in a:
  page = tag.get('href')
  if page not in unique_pages:
    print(page)
    food_url1 = 'https://foodista.com' + page

    # Use requests to retrieve data from a given URL
    food_response = requests.get(food_url1)
    dish=''
    # Parse the whole HTML page using BeautifulSoup
    food_soup = BeautifulSoup(food_response.text, 'html.parser')
    span = food_soup.find_all('span', {'class': 'field-content'})

    for ele in span:
      for value in ele.find_all('a'):
        dish=value.get('href')
        print(dish)
        recp.append(dish)

  t=food_soup.find_all('a', {'class': 'active'})
  for i in t:
    if i not in a:
      a.append(i)
```

```python
# Taking only the unique recipes
recp = list(set(recp))
```
